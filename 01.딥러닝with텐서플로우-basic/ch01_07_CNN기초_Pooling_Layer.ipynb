## Pooling layer
---

풀링(Pooling) 연산은 커널과 스트라이드 개념이 존재한다는 점에서 합성곱 연산과 유사하지만, 학습해야 할 가중치가 없으며 연산 후에 채널 수가 변하지 않습니다. 일반적으로 합성곱 층(합성곱 연산 + 활성화 함수) 다음에는 풀링 층을 추가하는 것이 일반적입니다. 


````python
from tensorflow.keras.layers import MaxPool2D, AvgPool2D
````

### **| 맥스 풀링(Max Pooling)**
맥스풀링은 이미지 영역의 최대값 만을 출력값으로 가지는 연산 방법입니다. 



<p align='center'>

````python
x = np.array([1,0,1,1,1,2,0,0,1,0,1,0,0,0,1,1], dtype = 'float16').reshape(1,4,4,1)
print(x[0,:,:,0],x.shape)
y = MaxPool2D(pool_size = 2, strides = 1)(x) # kernel_size = pool_size, input shape = (n,w,h,ch)
print(y[0,:,:,0], y.shape)
````

### **| 평균 풀링(Average Pooling)**


평균풀링은 이미지 영역의 평균값 을 출력값으로 가지는 연산 방법입니다.

## LeNet-5
---

LeNet-5[(논문)](http://yann.lecun.com/exdb/publis/pdf/lecun-01a.pdf) 구조는 아마도 가장 널리 알려진 CNN 구조일 것입니다. 이 구조는 1998년 얀 르쿤이 만들었으며 손글씨 숫자 인식(MNIST)에 널리 사용되었습니다.

### | LeNet-5을 이용한 손글씨 이미지 분류

####  **데이터 준비**

````python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# 전처리-정규화
x_train, x_test = x_train/255.0 , x_test/255.0
# 전처리-shape 맞추기
x_train = np.expand_dims(x_train, 3)
x_test = np.expand_dims(x_test, 3)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
````

#### **모델 구성**

````python
from tensorflow.keras import utils, layers, models
def build_lenet5() :
  x = layers.Input(shape=(28,28,1))
  y = layers.Conv2D(filters = 6, kernel_size = 5 ,padding='same', activation = 'tanh')(x)
  y = layers.AvgPool2D(pool_size = 2, strides = 2)(y) # pooling 레이어에서는 filters 없어서 채널 조정 불가
  y = layers.Conv2D(filters = 16,kernel_size = 5, padding = 'valid', activation = 'tanh')(y)
  y = layers.MaxPool2D(pool_size = 2, strides=2)(y)

  y = layers.Flatten()(y)
  y = layers.Dense(120,activation = 'tanh')(y)
  y = layers.Dense(84,activation = 'tanh')(y)
  y = layers.Dense(10,activation = 'softmax')(y)
  model = tf.keras.models.Model(x,y)
  return model
````

#### **학습 및 평가**

````python
model = build_lenet5()
model.summary()
model.compile(optimizer = 'sgd', loss = 'sparse_categorical_crossentropy', # 경사하강법 (배치개수에 한번씩 업데이트)
              metrics = ['accuracy']) 
model.fit(x_train, y_train, epochs=10)
print('='*30, '테스트스코어', '='*30)
model.evaluate(x_test, y_test)
````

### **| Feature map**

#### **feature map 얻기**

````python
layer_outputs = [layer.output for layer in model.layers]
model_output_layer = models.Model(inputs=model.input, outputs=layer_outputs) # 모든 레이어의 아웃풋 획득 가능
print(len(layer_outputs))
x_pred_maps = model_output_layer(x_test)
for x_pred_map in x_pred_maps:
  print(x_pred_map.shape)
cnn_layers = [1,2,3,4]
dnn_layers = [5,6,7,8]
````

#### **시각화**

![](https://www.datasciencecentral.com/wp-content/uploads/2021/10/1lvvWF48t7cyRWqct13eU0w.jpeg)

[시각화 함수 스크립트](https://github.com/yebiny/SkillTreePython-DeepLearning/blob/main/01.%EB%94%A5%EB%9F%AC%EB%8B%9Dwith%ED%85%90%EC%84%9C%ED%94%8C%EB%A1%9C%EC%9A%B0-basic/scripts/featuremap_dict.py)

````python
def show_featuremap(sample_idx, fmaps, cnn_idxs, flatten_idxs): # # sample_idx : 원하는 샘플의 인덱스, fmaps : 특징들이 담긴 정보 (convolution + pooling = fmap)
  for idx in cnn_idxs:                                          # cnn_idxs : convolution 인덱스, dnn_idxs : dense 인덱스
    fmap = fmaps[idx][sample_idx]
    chs = fmap.shape[2]
    print(f'{idx} 번째 Convolutional 레이어 {fmap.shape}')
    for ch in range(chs):
      plt.subplot(1,chs,ch+1)
      plt.imshow(fmap[:,:,ch], cmap='gray')
      plt.xticks([]);plt.yticks([])
    plt.show()

  for idx in flatten_idxs:
    fmap = fmaps[idx][sample_idx]
    print(f'{idx} 번째 Dense 레이어 {fmap.shape}')
    plt.figure(figsize=(int(len(fmap)/5),10))
    plt.imshow(tf.reshape(fmap,(1,len(fmap))), cmap='gray')
    plt.xticks(range(10));plt.yticks([]);plt.show()
````

````python
sample_idx = 38
show_featuremap(sample_idx, x_pred_maps, cnn_layers, dnn_layers)
````

## 과제 1. 의류 이미지 분류
---
* 데이터셋 : f-minst(fashoin mnist)
* 데이터를 소개하세요.

* 적절한 모델을 생성하세요

````python
mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print('기존 데이터 형태')
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
for i in range(10) :
  plt.subplot(2,5,i+1)
  plt.imshow(x_train[i])
plt.show()
# 전처리-정규화
x_train, x_test = x_train/255.0 , x_test/255.0

# 전처리-shape 맞추기
x_train = np.expand_dims(x_train, 3)
x_test = np.expand_dims(x_test, 3)
print('변형한 데이터 형태')
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
````

````python
print(set(y_train)) # 분류 가능한 클래스 개수 확인 (총 10가지)
'''
28X28 픽셀의 이미지 7만개
운동화, 셔츠, 샌들 등의 10가지로 분류될 수 있는 이미지.
'''
````

````python
from tensorflow.keras import utils, layers, models
def build_model() :
  x = layers.Input(shape=(28,28,1))
  y = layers.Conv2D(filters = 8, kernel_size = 5 ,padding='same', activation = 'relu')(x)
  y = layers.AvgPool2D(pool_size = 2, strides = 2)(y) 
  y = layers.Conv2D(filters = 16, kernel_size = 5 ,padding='same', activation = 'relu')(x)
  y = layers.AvgPool2D(pool_size = 2, strides = 2)(y) 
  y = layers.Conv2D(filters = 32, kernel_size = 5 ,padding='same', activation = 'relu')(x)
  y = layers.AvgPool2D(pool_size = 2, strides = 2)(y) 
  y = layers.Conv2D(filters = 64,kernel_size = 5, padding = 'valid', activation = 'relu')(y)
  y = layers.MaxPool2D(pool_size = 2, strides=2)(y)

  y = layers.Flatten()(y)
  y = layers.Dense(120,activation = 'relu')(y)
  y = layers.Dense(84,activation = 'relu')(y)
  y = layers.Dense(10,activation = 'softmax')(y)
  model = tf.keras.models.Model(x,y)
  return model
````

````python
model = build_model()
model.summary()
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', # 경사하강법 (배치개수에 한번씩 업데이트)
              metrics = ['accuracy']) 
model.fit(x_train, y_train, epochs=10)
print('='*30, '테스트스코어', '='*30)
model.evaluate(x_test, y_test)
````
