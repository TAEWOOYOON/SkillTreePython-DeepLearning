```python
# 3번 문제
a = np.linspace(0,205,100).reshape(100,1)
bkg = np.array([a for i in range(100)]).reshape(100,100).T
cv2_imshow(bkg)
```

```python
_, dst = cv2.threshold(bkg ,100, 255, cv2.THRESH_BINARY)
cv2_imshow(dst)
```

```python
# 7번 문제
c = (0,0,0)
s = (100,100,3)
bkg = np.full(s, 255, np.uint8)
bkg1 = np.full(s, 0, np.uint8)
cv2.circle(bkg, (50,50), 50, (0,255,0), -1)
cv2_imshow(bkg)
cv2.rectangle(bkg1, (0,0), (50,50), (255,255,255), -1)
cv2_imshow(bkg1)
result = cv2.bitwise_and(bkg,bkg1)
cv2_imshow(result)
```
