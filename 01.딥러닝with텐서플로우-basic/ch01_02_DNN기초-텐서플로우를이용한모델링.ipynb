#### **모델 구현**

```python
# 모든 레이어는 tensorflow.keras에 있음
from tensorflow.keras import layers, models
x = layers.Input(shape = (3), name = 'input_layer') # X : 데이터 개수,3)
y = layers.Dense(1, name = 'output_layer')(x) # Linear W, b Dense(output_shape)

model1 = models.Model(inputs=x, outputs=y, name = 'Model_1')
```

* 모델 시각화
  * `.summary()`
  * `tensorflow.keras.utils.plot_model()`

```python
# Output shape : 데이터 변환 과정
# None : 데이터 개수
# Params : 파라미터 개수
model1.summary()

from tensorflow.keras import utils
utils.plot_model(model1, show_shapes=True)

# Trainable param : 학습 가능한 파라미터 (역전파 통해서 업데이트)
# Non-trainable param : 학습 하지 않는 파라미터
```

#### **파라미터**

</p>


* 데이터 플로우 : 데이터의 변환 과정을 확인하세요.
  * `layer.output`


```python
for layer in model1.layers :
  print(layer.name) # 각 레이어의 이름
  print(layer.output.shape) # 각 레이어의 아웃풋
```

* 학습 파라미터 : 학습 파라미터의 shape를 확인하세요.
  * `.get_weights()`

```python
for params in model1.get_weights() :
  print(params.shape)

print(model1.get_weights())
```

#### **순전파 계산**

현재 초기화되어 있는 학습파라미터를 이용해 Y를 계산해 보세요.

* `np.random.normal`

```python
X = np.random.normal(size = (1,3))
print(X)
Y_pred = model1.predict(X)
print(Y_pred)
```

#### **모델 구현**

```python
from tensorflow.keras import layers, models, utils
x = layers.Input(shape=(4,))
z1 = layers.Dense(8)(x)
z2 = layers.Dense(8)(z1)
y = layers.Dense(3)(z2)
model4 = models.Model(inputs = x, outputs = y, name = 'Model_4')

model4.summary()
utils.plot_model(model4, show_shapes = True)
```

```python
'''
X : (N,4) N은 데이터 개수
Z1 : (N,8)
Z2 : (N,8)
Y : (N,3)
'''

for layer in model4.layers :
  print(layer.output.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import models, layers, utils
import numpy as np
tf.__version__
```

아래 그림과 같은 모델을 텐서플로우를 이용해서 만들어 봅시다.

### | Keras Functional API

* 모델 구현

```python
# 이 함수를 실행하면 아래와 같은 구성을 가진 모델이 반환
def build_model() :
  x = layers.Input(shape = (3,))
  z1 = layers.Dense(6)(x)
  z2 = layers.Dense(6)(z1)
  y = layers.Dense(1)(z2)
  model = models.Model(x,y,)
  return model
```

* 시각화

```python
model = build_model()
model.summary()
utils.plot_model(model, show_shapes=True)
```

### | Keras Sequential API



* 모델 구현: `Sequential()`

```python
def build_model2() :
  model = tf.keras.Sequential() # 빈 껍질을 만들어 놓은것
  model.add( layers.Input(shape = (3)) )
  model.add( layers.Dense(shape = (6)) )
  model.add( layers.Dense(shape = (6)) )
  model.add( layers.Dense(shape = (1)) )
  return model
```

* 모델 시각화

```python
m2 = build_model()
m2.summary()
utils.plot_model(m2, show_shapes=True)
```

### | Model Class Subclassing

* 모델 구현

```python
class Model3(tf.keras.Model) : # 이미 존재하는 클래스를 불러올 때 괄호 안에 값 입력
  def __init__(self) :
    super(Model3, self).__init__() # tf.keras.Model에 이미 정의되어 있는 파라미터 불러오기
    self._name = 'Model3'
    self._input_shape=(3)
    self.dense1 = layers.Dense(6, name = 'dense1')
    self.dense2 = layers.Dense(6, name = 'dense2')
    self.dense3 = layers.Dense(3, name = 'dense3')

  def call(self, x) :
    y = self.dense1(x)
    y = self.dense2(x)
    y = self.dense3(x)
    return y
```

* 모델 시각화

```python
m3 = Model3()
m3(np.array([[1,2,1]]))
m3.summary()
utils.plot_model(m3)
```
